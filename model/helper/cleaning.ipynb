```python
def combine(folder,drop=None,rename_dict=None):
    files = os.listdir(folder)

    df_list = []
    i = 0
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(folder, file)
            df = pd.read_csv(file_path)
            df_list.append(df)
            i+=1
            print(i)

    combined_df = pd.concat(df_list, ignore_index=True)
    combined_df = combined_df.drop_duplicates()
    combined_df['date'] = pd.to_datetime(combined_df['date']).dt.day
    if drop:
        combined_df.drop(drop,axis=1,inplace=True)
    if rename_dict:
        combined_df.rename(columns=rename_dict, inplace=True)
    combined_df.fillna(0.0,inplace=True)
    combined_df.to_csv(f'{folder}/combined.csv', index=False)

rename_dict = {'_state_': 'State', 'district': 'District'}
drop = ['basin','subbasin']
combine('D:\projects\water_forecasting\data/rainfall',drop,rename_dict)
```

```python
import pandas as pd

df = pd.read_csv('data/usage/Usage.csv')
df['note'] = df['note'].fillna('')
df['Irrigation'] = df.apply(lambda row: row['value'] if row['purpose'] == 'Irrigation' else 0, axis=1)
df['Industrial'] = df.apply(lambda row: row['value'] if row['purpose'] == 'Industrial' and 'Includes domestic also' not in row['note'] else 0, axis=1)
df['Domestic'] = df.apply(lambda row: row['value'] if row['purpose'] == 'Domestic' and 'Includes domestic also' not in row['note'] else 0, axis=1) 

df.loc[df['note'].str.contains('Includes domestic also', na=False), 'Industrial'] = df['value'] / 2
df.loc[df['note'].str.contains('Includes domestic also', na=False), 'Domestic'] = df['value'] / 2

df_pivot = df.pivot_table(index=['year', 'state'], 
                          values=['Irrigation', 'Industrial', 'Domestic'], 
                          aggfunc='sum').reset_index()

df_pivot['Total'] = df_pivot['Irrigation'] + df_pivot['Industrial'] + df_pivot['Domestic']

df_pivot.to_csv('data/usage.csv', index=False)

print(df_pivot.head())
```

```python
import os
import pandas as pd

root_dir = 'data/luc'  # Use forward slashes or os.path.join for compatibility

combined_data = []

for year_folder in os.listdir(root_dir):
    year_path = os.path.join(root_dir, year_folder)
    
    if os.path.isdir(year_path):
        # Clean year folder name
        year = year_folder.strip().replace('.', '')
        
        for file in os.listdir(year_path):
            if file.endswith('.csv'):
                file_path = os.path.join(year_path, file)
                
                # Extract the state name from the filename
                state = file.split('_')[0]
                
                try:
                    df = pd.read_csv(file_path)
                    
                    # Add State and Year columns
                    df['State'] = state
                    df['Year'] = year
                    
                    # Reorder the columns to place 'State' and 'Year' at the beginning
                    cols = ['State', 'Year'] + [col for col in df.columns if col not in ['State', 'Year']]
                    df = df[cols]
                    
                    # Append only if the DataFrame is not empty
                    if not df.empty:
                        combined_data.append(df)
                        print(f"Successfully processed: {file_path}")
                
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if combined_data:
    final_df = pd.concat(combined_data, ignore_index=True)
    
    # Dynamically generate output file name based on processed years
    output_file_name = f"Combined_Land_Use_Data.csv"
    output_path = os.path.join(root_dir, output_file_name)
    
    final_df.to_csv(output_path, index=False)
    
    print(f"\nAll files have been successfully combined into {output_path}")
else:
    print("No data to combine.")
```

```python
import os
import pandas as pd

base_path = 'data/luc'

expected_columns = [
    'Column1', 'Column2', 'Column3', 'Column4', 'Column5', 'Column6', 'Column7',
    'Column8', 'Column9', 'Column10', 'Column11', 'Column12', 'Column13', 
    'Column14', 'Column15', 'Column16', 'Column17'
]

def clean_dataframe(df):
    # Remove completely empty rows and reset index
    df.dropna(how='all', inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    # Remove the first 4 rows (assumed to be metadata or headers)
    df = df.iloc[4:]
    df.replace('-', 0, inplace=True)
    df.reset_index(drop=True, inplace=True)

    # Check if the dataframe has at least 17 columns and extract the "Total" row
    if not df.empty:
        if len(df.columns) >= 17:
            # Extract the last row which is assumed to be the "Total" row
            df_total = df.iloc[[-1], [3, 6, 10, 13, 15]]  # Selecting the total row for specific columns
            df_total.columns = ['Forest', 'Barren', 'Others', 'Fallow', 'Cropped']
            return df_total
        else:
            print("File does not contain the expected 17 columns, skipping this file.")
            return None

    return None

all_data = []

# Traverse through all the year folders and CSV files
for year_folder in os.listdir(base_path):
    year_folder_path = os.path.join(base_path, year_folder)
    
    if os.path.isdir(year_folder_path):
        year = year_folder.replace('-', '/')
        
        for file in os.listdir(year_folder_path):
            if file.endswith('.csv'):
                state = file.split('_')[0]
                year = year.split('/')[0]
                file_path = os.path.join(year_folder_path, file)
                
                try:
                    df = pd.read_csv(file_path)
                    df_cleaned = clean_dataframe(df)
                    
                    if df_cleaned is not None:
                        df_cleaned.insert(0, 'State', state)  # Add state to the first column
                        df_cleaned.insert(1, 'Year', year)    # Add year to the second column
                        all_data.append(df_cleaned)

                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")

if all_data:
    # Combine all the "Total" rows into one DataFrame
    combined_df = pd.concat(all_data, ignore_index=True)

    # Save the combined "Total" rows to a CSV file
    output_file = 'data/luc.csv'
    combined_df.to_csv(output_file, index=False)
    print(f"Data combined and saved to {output_file}")
else:
    print("No data to combine.")
```

```python
import os
import pandas as pd

base_path = 'data/evapuration'

all_data = []

# Function to split the date into day, month, and year
def split_date_column(df,file):
    if 'date' in df.columns:
        # Handle dates with both '-' and '/' separators
        if df['date'].str.contains('/').all():
            try:
                df[['year', 'month', 'day']] = df['date'].str.split('/', expand=True)
                df.drop(columns=['date'], inplace=True)
            except ValueError as ve:
                print(f"Error splitting date in dataframe: {ve}")
                return None
        elif df['date'].str.contains('-').all():
            try:
                df[['year', 'month', 'day']] = df['date'].str.split('-', expand=True)
                df.drop(columns=['date'], inplace=True)
            except ValueError as ve:
                print(f"Error splitting date in dataframe: {ve}")
                return None
        else:
            print(f"Date column does not contain expected separator ('-' or '/'), skipping {file}.")
            return None
    else:
        print("No date column found, skipping file.")
        return None
    return df

# Traverse through all the files and combine them
for file in os.listdir(base_path):
    if file.endswith('.csv'):
        file_path = os.path.join(base_path, file)
        
        try:
            df = pd.read_csv(file_path)
            df = split_date_column(df,file)

            # Only append if the dataframe was processed successfully
            if df is not None:
                all_data.append(df)

        except Exception as e:
            print(f"Error processing file {file}: {e}")

# Combine all dataframes into one if any were processed successfully
if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    output_file = 'data/evapuration/evapuration.csv'
    combined_df.to_csv(output_file, index=False)
    print(f"Data combined and saved to {output_file}")
else:
    print("No valid data to combine.")
```

```python
import os
import pandas as pd

base_path = 'data/luc'

# Function to clean and extract necessary data from the DataFrame
def clean_dataframe(df):
    # Remove empty rows and reset index
    df.dropna(how='all', inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    # Remove the first 4 rows (assumed to be metadata or headers)
    df = df.iloc[4:]
    df.reset_index(drop=True, inplace=True)

    # Replace '-' with 0 in the DataFrame
    #df.replace('-', 0, inplace=True)
    
    # Check if the DataFrame has at least 17 columns
    if len(df.columns) >= 17:
        # Extract the last row, which is assumed to be the "Total" row
        df_total = df.iloc[[-1], [3, 6, 10, 13, 15]]  # Selecting the total row for specific columns
        df_total.columns = ['Forest', 'Barren', 'Others', 'Fallow', 'Cropped']  # Removed District
        return df_total
    else:
        print("File does not contain the expected columns, skipping this file.")
        return None

all_data = []

# Traverse all year folders
for year_folder in os.listdir(base_path):
    year_folder_path = os.path.join(base_path, year_folder)
    
    if os.path.isdir(year_folder_path):
        # Extract only the first year (e.g., '2004' from '2004/2005')
        year = year_folder.split('-')[0]  # Get the first year from folder name
        
        # Traverse all CSV files in the year folder
        for file in os.listdir(year_folder_path):
            if file.endswith('.csv'):
                state = file.split('_')[0]  # Extract state from filename
                file_path = os.path.join(year_folder_path, file)
                
                try:
                    df = pd.read_csv(file_path)
                    # Replace '-' with 0 in the DataFrame right after reading
                    df.replace('-', 0, inplace=True)
                    df_cleaned = clean_dataframe(df)
                    
                    if df_cleaned is not None:
                        df_cleaned.insert(0, 'State', state)  # Add state to the first column
                        df_cleaned.insert(1, 'Year', year)    # Add year to the second column
                        all_data.append(df_cleaned)

                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")

if all_data:
    # Combine all the "Total" rows into one DataFrame
    combined_df = pd.concat(all_data, ignore_index=True)

    # Remove rows where any column has 'Total' in the row (if any remain)
    cleaned_combined_df = combined_df[~combined_df.astype(str).apply(lambda x: x.str.contains('Total', na=False)).any(axis=1)]
    
    # Convert all columns except 'State' and 'Year' to float
    for col in cleaned_combined_df.columns:
        if col not in ['State', 'Year']:
            cleaned_combined_df[col] = cleaned_combined_df[col].astype(float)

    # Calculate a new "Total" row by summing numeric columns
    numeric_columns = cleaned_combined_df.select_dtypes(include='number').columns
    total_row = cleaned_combined_df[numeric_columns].sum(numeric_only=True)
    
    # Create a DataFrame for the total row and add 'State' and 'Year' columns with 'Total'
    total_row_df = pd.DataFrame(total_row).transpose()
    total_row_df.insert(0, 'State', 'Total')
    total_row_df.insert(1, 'Year', 'Total')

    # Append the "Total" row to the combined DataFrame
    final_df = pd.concat([cleaned_combined_df.drop(columns=['District'], errors='ignore'), total_row_df], ignore_index=True)

    # Save the final DataFrame with the new "Total" row to CSV
    output_file = 'data/luc.csv'
    final_df.dropna(inplace=True)
    final_df.to_csv(output_file, index=False)
    
    print(f"Data combined and saved to {output_file}, including a new 'Total' row.")
else:
    print("No data to combine.")
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv('data/population/population.csv')

# Replace specified values with NaN
df.replace(to_replace=[0, '-', 'Null'], value=np.nan, inplace=True)

# Drop rows with NaN values only in population columns while keeping 'Year'
df.dropna(subset=['Population', 'Urban Population', 'Rural Population'], inplace=True)

# Columns to clean
population_columns = ['Population', 'Urban Population', 'Rural Population']

# Clean and convert population columns to float
for column in population_columns:
    df[column] = df[column].str.replace(',', '')  # Remove commas
    df[column] = df[column].astype(float)  # Convert to float

# Select only 'Year' and population columns for the final output
final_columns = ['Year'] + population_columns
final_df = df[final_columns]

# Save the cleaned DataFrame to a new CSV file
final_df.to_csv('data/population.csv', index=False)

print("Data cleaned and saved to 'data/population.csv'.")
```
